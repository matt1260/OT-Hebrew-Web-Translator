```python
def decode_morpheme(piece, lang):
    decoded = []
    
    # Initialize dictionaries
    noun_types = {
        'c': 'common',
        'g': 'gentilic',
        'p': 'proper name',
    }

    adjective_types = {
        'a': 'adjective',
        'c': 'cardinal number',
        'g': 'gentilic',
        'o': 'ordinal number',
    }

    pronoun_types = {
        'd': 'demonstrative',
        'f': 'indefinite',
        'i': 'interrogative',
        'p': 'personal',
        'r': 'relative',
    }

    preposition_types = {
        'd': 'definite article',
    }

    suffix_types = {
        'd': 'directional he',
        'h': 'paragogic he',
        'n': 'paragogic nun',
        'p': 'pronominal',
    }
    
    particle_types = {
        'a': 'affirmation',
        'd': 'Art',
        'e': 'exhortation',
        'i': 'interrogative',
        'j': 'interjection',
        'm': 'demonstrative',
        'n': 'negative',
        'o': 'DirObjM',
        'r': 'relative',
    }

    gender = {
        'b': 'both (noun)',
        'c': 'common (verb)',
        'f': 'feminine',
        'm': 'masculine',
    }
    
    person = {
        '1': 'first',
        '2': 'second',
        '3': 'third',
    }

    number = {
        'd': 'dual',
        'p': 'plural',
        's': 'singular',
    }

    state = {
        'a': 'absolute',
        'c': 'construct',
        'd': 'determined',
    }

    verb_stems_hebrew = {
    'q': 'Qal',
    'N': 'Niphal',
    'p': 'Piel',
    'P': 'Pual',
    'h': 'Hiphil',
    'H': 'Hophal',
    't': 'Hithpael',
    'o': 'Polel',
    'O': 'Polal',
    'r': 'Hithpolel',
    'm': 'Poel',
    'M': 'Poal',
    'k': 'Palel',
    'K': 'Pulal',
    'Q': 'Qal Passive',
    'l': 'Pilpel',
    'L': 'Polpal',
    'f': 'Hithpalpel',
    'D': 'Nithpael',
    'j': 'Pealal',
    'i': 'Pilel',
    'u': 'Hothpaal',
    'c': 'Tiphil',
    'v': 'Hishtaphel',
    'w': 'Nithpalel',
    'y': 'Nithpoel',
    'z': 'Hithpoel',
    }


    verb_conjugation_types_hebrew = {
        'p': 'perfect (qatal)',
        'q': 'sequential perfect (weqatal)',
        'i': 'imperfect (yiqtol)',
        'w': 'sequential imperfect (wayyiqtol)',
        'h': 'cohortative',
        'j': 'jussive',
        'v': 'imperative',
        'r': 'participle active',
        's': 'participle passive',
        'a': 'infinitive absolute',
        'c': 'infinitive construct',
    }

    verb_stems_aramaic = {
        'q': 'peal',
        'Q': 'peil',
        'u': 'hithpeel',
        'p': 'pael',
        'P': 'ithpaal',
        'M': 'hithpaal',
        'a': 'aphel',
        'h': 'haphel',
        's': 'saphel',
        'e': 'shaphel',
        'H': 'hophal',
        'i': 'ithpeel',
        't': 'hishtaphel',
        'v': 'ishtaphel',
        'w': 'hithaphel',
        'o': 'polel',
        'z': 'ithpoel',
        'r': 'hithpolel',
        'f': 'hithpalpel',
        'b': 'hephal',
        'c': 'tiphel',
        'm': 'poel',
        'l': 'palpel',
        'L': 'ithpalpel',
        'O': 'ithpolel',
        'G': 'ittaphal',
    }

    verb_conjugation_types_aramaic = {
        'p': 'perfect (qatal)',
        'q': 'sequential perfect (weqatal)',
        'i': 'imperfect (yiqtol)',
        'w': 'sequential imperfect (wayyiqtol)',
        'h': 'cohortative',
        'j': 'jussive',
        'v': 'imperative',
        'r': 'participle active',
        's': 'participle passive',
        'a': 'infinitive absolute',
        'c': 'infinitive construct',
    }

    # Determine the type of the piece based on the second letter
    
    # print('Morpheme:', piece)

    piece_type = piece[0]
    if piece[0] == 'H':
        piece_type = piece[1]
    if piece[0] == 'A':
        piece_type = piece[1]
        
    # print('Morpheme Type:',piece_type)
    
    if piece_type in 'Cc':
        decoded.append('Conj-w')

    if piece_type == 'N':
        decoded.append('noun')
        decoded.append(noun_types.get(piece[1], ''))
        for letter in piece[2:]:
            if letter in gender:
                decoded.append(gender[letter])
            elif letter in number:
                decoded.append(number[letter])
            elif letter in state:
                decoded.append(state[letter])

    elif piece_type == 'A':
        decoded.append('adjective')
        decoded.append(adjective_types.get(piece[1], ''))
        for letter in piece[2:]:
            if letter in gender:
                decoded.append(gender[letter])
            elif letter in number:
                decoded.append(number[letter])
            elif letter in state:
                decoded.append(state[letter])

    elif piece_type == 'P':
        decoded.append('pronoun')
        decoded.append(pronoun_types.get(piece[1], ''))
        for letter in piece[2:]:
            if letter in gender:
                decoded.append(gender[letter])
            elif letter in number:
                decoded.append(number[letter])
            elif letter in state:
                decoded.append(state[letter])

    elif piece_type == 'S':
        decoded.append('suffix')
        decoded.append(suffix_types.get(piece[1], ''))
        for letter in piece[2:]:
            if letter in person:
                decoded.append(person[letter])
            elif letter in gender:
                decoded.append(gender[letter])
            elif letter in number:
                decoded.append(number[letter])

    elif piece_type == 'R':
        decoded.append('preposition')

    elif piece_type == 'T':
        #decoded.append('')
        if piece[0] in ['HA']:
            particle_code = piece[1+1] # if it is the first piece
        else: particle_code = piece[1]
                print(piece[1])
                
        for letter in piece[2:]:
            if letter in particle_types:
                decoded.append(particle_types[letter])

        #decoded.append(particle_types.get(particle_code, ''))
        
    elif piece_type == 'V':
        decoded.append('verb')
        if lang == 'H': # Hebrew Verbs
            decoded.append(verb_stems_hebrew.get(piece[2], ''))
            if len(piece) == 6:
                decoded.append(verb_conjugation_types_hebrew.get(piece[2], ''))
                for letter in piece[3:]:
                    if letter in person:
                        decoded.append(person[letter])
                    elif letter in gender:
                        decoded.append(gender[letter])
                    elif letter in number:
                        decoded.append(number[letter])
            else:
            #     decoded.append(verb_conjugation_types_hebrew.get(piece[2], ''))

                for letter in piece[4:]:
                    if letter in person:
                        decoded.append(person[letter])
                    elif letter in gender:
                        decoded.append(gender[letter])
                    elif letter in number:
                        decoded.append(number[letter])


        elif lang == 'A': # Aramaic Verbs
            decoded.append(verb_stems_aramaic.get(piece[2], ''))
            if len(piece) == 6:
                decoded.append(verb_conjugation_types_aramaic.get(piece[3], ''))
            else:
                decoded.append(verb_conjugation_types_aramaic.get(piece[2], ''))
            for letter in piece[4:]:
                if letter in person:
                    decoded.append(person[letter])
                elif letter in gender:
                    decoded.append(gender[letter])
                elif letter in number:
                    decoded.append(number[letter])
    
    # print(decoded)
    
    return '|'.join(decoded)

def decode_code(code):
    
    if code and '/' in code:
        pieces = code.split('/')
        decoded_morphemes = [decode_morpheme(piece, 'H') if code[0] == 'H' else decode_morpheme(piece, 'A') for piece in pieces]
        return ' - '.join(decoded_morphemes)

    else:
        lang = 'H' if code[0] == 'H' else 'A'
        return decode_morpheme(code, lang)

# Example usage:
code = 'HTc'
decoded_result = decode_code(code)
print(f'{code}: {decoded_result}')
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('rbt_hebrew.sqlite3')
cursor = conn.cursor()

# Run the SQL query to count matches and list them with their counts
query = """
    SELECT Morph, COUNT(*) as Count
    FROM hebrewdata
    GROUP BY Morph
    ORDER BY Count DESC;
"""

cursor.execute(query)

# Fetch and print the results
results = cursor.fetchall()
for row in results:
    morph, count = row
    if morph is not None:

        decoded_morph = decode_code(morph)  # Decode the 'Morph' value
        print(f'Morph: {morph} - {decoded_morph}, Count: {count}')
    else:
        print(f'Morph: {morph} - Count: {count}')

# Close the database connection
conn.close()
```
